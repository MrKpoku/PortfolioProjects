```python
# Import Libraries

import pandas as pd
import seaborn as sns


import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration fo the plots we will create

# Read in the data
df = pd.read_csv(r"C:\Users\kingk\Downloads\movies.csv\movies.csv")
```

```python
# Let's Look at the data

df.head()
```

```python
# Data types for the columns
df.dtypes
```

```python
#Change data type of columns

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
df['votes'] = df['votes'].fillna(0).astype('int64')
```

```python
df
df.head()
```

```python
#Sorting data according to gross revenue
#the shinning,the blue lagoon and star wars are the top 3 movies that generated the most gross revenue with the united states repeating twice in the top 3

df=df.sort_values(by=['gross'],inplace=False,ascending=False)
df.head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Removing any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
df.head()
```

```python
df.head()
```

```python
# Budget High correlation
#company will have a high correlation - these are my educated guess but lets test them
```

```python
# Building a Scatter plot with buget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Revenue')

plt.xlabel('Gross Revenue')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plot the budget vs gross revenue using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "grey"}, line_kws={"color": "yellow"})
plt.show()
```

```python
# Looking into the correlation 

numeric_columns = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_columns].corr() #pearson 

# Display the correlation matrix
correlation_matrix
```

```python
# High correlation between budget and gross with a whopping 0.750157(that is pretty high)
```

```python
correlation_matrix = df[numeric_columns].corr()

sns.heatmap(correlation_matrix,annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
